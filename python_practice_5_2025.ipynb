5. (9б). Створіть клас 3D-фігура без атрибутів, з методом get_volume для отримання об'єму фігури, що повертає 0 та __add __, який приймає self та other в якості аргументів, а повертає суму об'ємів 3d-фігур self та other.

```python
class Figure3d:
  def get_volume(self):
    return 0

  def __add__(self, other):
    return self.get_volume() + other.get_volume()
```

6. (11б). Створіть 2 дочірніх класи від 3d-фігура: Циліндр та Куб, які мають атрибути, необхідні для розрахунку об'єму. Визначте метод get_volume в кожному з них так, щоби вони розраховували об'єм у залежності від формули для кожного типу фігури. Створіть обʼєкт класу Циліндр та обʼєкт класу Куб. Виконайте операцію суми за допомогою оператора + між ними, аби знайти сумарний об'єм.

```python
class Cube(Figure3d):
  def __init__(self, side):
    self.side = side

  def get_volume(self):
    return self.side ** 3

class Cylinder(Figure3d):
  def __init__(self, height, radius):
    self.height = height
    self.radius = radius

  def get_volume(self):
    return 3.14 * (self.radius ** 2) * self.height

cube = Cube(side = 3)
cylinder = Cylinder(height = 3, radius = 3)

sum_volume = cube + cylinder

print(cube.get_volume(), cylinder.get_volume(), sum_volume)
```

8. (13б). Створіть клас EWallet з приватними атрибутами balance та id. Реалізуйте методи поповнення та зняття коштів, забезпечивши належну інкапсуляцію. Підказка: використовуйте декоратори property та setter.

```python
class EWallet:
  def __init__(self, id, balance = 0.0):
    self.__balance = balance
    self.__id = id

  @property
  def balance(self):
    return self.__balance

  @balance.setter
  def balance(self, value):
    if value < 0:
      return "balance cannot be less than 0"
    else:
      self.__balance = value

  @property
  def id(self):
    return self.__id

  def deposit(self, amount):
    if amount > 0:
      self.balance += amount
    else:
      return "amount to deposit must be more than 0"

  def withdraw(self, amount):
    if 0 < amount <= self.balance:
      self.balance -= amount
    else:
      return "you do not have enough money"

wallet = EWallet("123456", 150.0)

wallet.deposit(50)
wallet.withdraw(100)

print(wallet.balance)
```

9. (11б). Створіть клас ShoppingCart, який містить список об'єктів типу Item. Реалізуйте методи для додавання айтемів, їх видалення та відображення списку айтемів.

```python
class Item:
  def __init__(self, name):
    self.name = name

  def __str__(self):
        return f"name: {self.name}"

class ShoppingCart:
  def __init__(self):
    self.items = []

  def add_item(self, item):
    if isinstance(item, Item):
      self.items.append(item)
    else:
      return "you cannot add anything except item"

  def delete_item(self, item_name):
    for item in self.items:
      if item.name == item_name:
        self.items.remove(item)
        return
    return "item was not found"

  def item_list(self):
    if not self.items:
      return "cart is empty"
    else:
      for index, item in enumerate(self.items, 1):
        print(f"{index}. {item}")

apple = Item("apple")
cart = ShoppingCart()

cart.add_item(apple)
cart.item_list()
```

10. (13б). Створіть клас Vehicle з атрибутами brand (manufacturer) та production_age. Створіть ще один клас Engine з такими атрибутами, як type (ex. petrol) та size (volume, litres). Створіть клас Car, який успадковує обидва класи Vehicle та Engine. Продемонструйте використання множинної спадковості, створивши обʼєкт класу Car та отримавши mro для цього класу.

```python
class Vechicle:
  def __init__(self, brand, production_age):
    self.brand = brand
    self.production_age = production_age

class Engine:
  def __init__(self, typeP, size):
    self.typeP = typeP
    self.size = size

class Car(Vechicle, Engine):
  def __init__(self, brand, production_age, typeP, size):
    Vechicle.__init__(self, brand, production_age)
    Engine.__init__(self, typeP, size)

  def get_info(self):
    return f"{self.brand}, {self.production_age}, {self.typeP}, {self.size}"

car = Car("BMW", 2020, "Petrol", 3.0)
print(car.get_info())

print("\nMRO")
for cls in Car.mro():
    print(cls)
```

13. (11б). Створіть клас, що має визначений Iterator Protocol та при кожному виклику next повертає назву дня тижня, поки вони не вичерпаються. Зверніть увагу, що порядок має значення.

```python
class Week:
  def __init__(self):
    self.days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
    self.index = 0

  def __iter__(self):
    return self

  def __next__(self):
    if self.index < len(self.days):
      day = self.days[self.index]
      self.index += 1
      return day
    else:
      raise StopIteration

week = Week()

for day in week:
    print(day)
```
